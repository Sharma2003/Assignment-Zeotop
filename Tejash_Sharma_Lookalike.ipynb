```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
customer = pd.read_csv("Dataset/Customers.csv")
transactions = pd.read_csv("Dataset/Transactions.csv")
products = pd.read_csv("Dataset/Products.csv")
```

```python
df = transactions.merge(customer, on="CustomerID", how="left")
df = df.merge(products, on="ProductID", how="left")
```

```python
df.head()
```

```python
customer_product_matrix = df.pivot_table(index="CustomerID", columns="ProductID", values="Quantity", aggfunc="sum").fillna(0)
```

```python
customer_product_matrix
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(customer_product_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(scaled_features)
```

```python
customer_ids = customer_product_matrix.index
similarity_df = pd.DataFrame(similarity_matrix, index=customer_ids, columns=customer_ids)
```

```python
similarity_df
```

```python
def get_similar_customers(customer_id):
    if customer_id not in similarity_df.index:
        return "Customer ID not found!"
    
    similar_customers = similarity_df[customer_id].sort_values(ascending=False).index[1:4]  # Exclude self, get top 3
    results = [(cust_id, similarity_df.at[customer_id, cust_id]) for cust_id in similar_customers]
    
    return results
```

```python
user_id = input("Enter Customer ID (e.g., C0001): ")
recommendations = get_similar_customers(user_id)
print(f"\nTop 3 similar customers for {user_id}:")
for cust, score in recommendations:
    print(f"- {cust} (Similarity Score: {score:.2f})")
```

```python
df = pd.DataFrame(recommendations, columns=['Customer ID', 'Similarity Score'])
```

```python
df.to_csv("Lookalike.csv", index=False)
```
